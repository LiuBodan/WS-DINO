```python
import pandas as pd

df = pd.read_csv('BBBC021_annotated_and_DMSO_illumCorrected(2329).csv')

# Extract columns for wanted ones
sub_df = df[["Image_FileName_DAPI","Image_PathName_DAPI","Image_FileName_Tubulin","Image_PathName_Tubulin","Image_FileName_Actin","Image_PathName_Actin","Image_Metadata_Plate_DAPI"]]

pd.set_option('display.max_colwidth', None)
```

```python
# Modify the entries of these columns

for k in sub_df.keys():
    sub_df.loc[:,k] = sub_df.loc[:,k].str.replace('_corrected_resized.tiff', '.tif')
    sub_df.loc[:,k] = sub_df.loc[:,k].str.replace('/home/qian.wang/WS-DINO/preprocessing/outCellProfiler/illumCorrectedImages/', '')


sub_df = sub_df.rename(columns={'Image_Metadata_Plate_DAPI': 'Metadata_Plate'})
```

Change the fodler path to fit the csv file

```python
import os
import shutil

source_path = "D:/archive"  # Replace with your source directory path
destination_path = os.path.join("D:/BBBC021/", "Week9")

# Create the new folder if it doesn't exist
if not os.path.exists(destination_path):
    os.makedirs(destination_path)

# Loop through items in the source path
for item in os.listdir(source_path):
    item_path = os.path.join(source_path, item)
    
    # Check if the item is a directory and contains the string 'Week1'
    if os.path.isdir(item_path) and 'Week9_' in item:
        for subfolder in os.listdir(item_path):
            subfolder_path = os.path.join(item_path, subfolder)
            
            # Check if it's a directory
            if os.path.isdir(subfolder_path):
                # Move the subfolder to the destination path
                shutil.move(subfolder_path, destination_path)
```
